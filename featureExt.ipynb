```python
eastern=pd.read_excel(r"C:\work\Geo Task\Eaestern half data.xlsx")
```

```python
eastern['State']=""
eastern['state_district']=""
eastern['county']=""
eastern['city']=""
eastern['postcode']=""
eastern['suburb']=""
for i in eastern.index:
    if eastern.loc[i,'Resp']!="":
            try:
                eastern.loc[i,'State']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['state']
            except:
                 pass
            try:
                eastern.loc[i,'state_district']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['state_district']
            except:
                 try:
                    eastern.loc[i,'state_district']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['district']
                 except:
                      pass
            try:
                eastern.loc[i,'county']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['county']
            except:
                 pass
            try:
                eastern.loc[i,'city']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['city']
            except:
                 pass
            try:
                eastern.loc[i,'postcode']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['postcode']
            except:
                 pass
            try:
                eastern.loc[i,'suburb']=ast.literal_eval(eastern['Resp'][i])['features'][0]['properties']['suburb']
            except:
                 pass
            
```
