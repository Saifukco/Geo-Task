```python
def ger_resp(x):
    l=x.split(',')
    lat=l[0]
    lon=l[1]
    url = "https://api.geoapify.com/v1/geocode/reverse?lat={}&lon={}&apiKey=e79fb11fadfc46a4ad44b6151ecdad75".format(lat,lon)
    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    resp = requests.get(url, headers=headers)
    return str(resp.json())
```

```python
def ger_respSahana(x):
    l=x.split(',')
    lat=l[0]
    lon=l[1]
    url = "https://api.geoapify.com/v1/geocode/reverse?lat={}&lon={}&apiKey=eb6162ced3544f43a1338edb9e7039a6".format(lat,lon)
    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    resp = requests.get(url, headers=headers)
    return str(resp.json())
```

```python
def ger_respShrisha(x):
    l=x.split(',')
    lat=l[0]
    lon=l[1]
    url = "https://api.geoapify.com/v1/geocode/reverse?lat={}&lon={}&apiKey=63f053f4591140239b42cec2c70b3529".format(lat,lon)
    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    resp = requests.get(url, headers=headers)
    return str(resp.json())
```

```python
def ger_respArya(x):
    l=x.split(',')
    lat=l[0]
    lon=l[1]
    url = "https://api.geoapify.com/v1/geocode/reverse?lat={}&lon={}&apiKey=422ac9ee85714ca1ab53df77488777f8".format(lat,lon)
    headers = CaseInsensitiveDict()
    headers["Accept"] = "application/json"
    resp = requests.get(url, headers=headers)
    return str(resp.json())
```
